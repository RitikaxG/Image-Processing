```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Loading an Image using OpenCV

```python
img = cv.imread('Transition.jpg')
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Extracting using slicing
img[:1000,:,:].shape
```

### Frame Extraction
* Extracting red, green and blue frames from an RGB image

```python
# 300*400*3
r = img[:,:,0] # Red
g = img[:,:,1] # Green
b = img[:,:,2] # Blue
```

```python
plt.imshow(r)
```

```python
plt.imshow(b)
```

```python
plt.imshow(g)
```

```python
cv.imwrite('Blue.jpg',b)
cv.imwrite('Red.jpg',r)
cv.imwrite('Green.jpg',g)
```
